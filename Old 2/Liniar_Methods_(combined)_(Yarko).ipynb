```
import numpy as np
import pandas as pd
from interval import interval, inf, imath
from sympy.solvers import solve
from sympy import Symbol
```

```
def linear_half(f, start, end, e):
  check = end - start
  if check <= e:
    print(f"Found x = [{start}; {end}]")
    return

  X0 = interval([start, end])
  mid = (start+end)/2;
  f_X0 = f(X0)

  intervals = np.array([f"[{start}; {end}]"])
  widths = np.array([check])
  middles = np.array([mid])
  function_extensions = np.array([f"[{f_X0[0].inf}; {f_X0[0].sup}]"])

  if 0 in f_X0:
    intervals_stack = [(mid, end), (start, mid)]
    while len(intervals_stack) != 0:
      Xn = interval(intervals_stack.pop())
      start = Xn[0].inf
      end = Xn[0].sup
      check = end - start
      if check <= e:
        #print('FOUND X =', start, ' ', end)
        intervals = np.append(intervals, "Result")
        widths = np.append(widths, None)
        middles = np.append(middles, None)
        function_extensions = np.append(function_extensions, f"Found x: [{start}; {end}]")
        #skip the rest, we found one on this interval
        continue

      mid = (start + end) / 2;
      f_Xn = f(Xn)

      intervals = np.append(intervals, f"[{start}, {end}]")
      widths = np.append(widths, check)
      middles = np.append(middles, mid)
      function_extensions = np.append(function_extensions, f"[{f_Xn[0].inf}, {f_Xn[0].sup}]")

      if 0 in f_Xn:
        intervals_stack.append((mid, end))
        intervals_stack.append((start, mid))
      else:
        continue

    #display pandas table with results as how they were processed
    #intervals = np.insert(intervals, 0, "Xn")
    #widths = np.insert(widths, 0, "ω(Xn)")
    #middles = np.insert(middles, 0, "x̅_n")
    #function_extensions = np.insert(function_extensions, 0, "F(Xn)")
    pd.set_option('display.max_rows', 999)  
    df = pd.DataFrame({"Xn": intervals, "ω(Xn)": widths, "x̅_n": middles, "F(Xn)": function_extensions})
    display(df)
    return
  
#@title Поділ навпіл (задача)
print("Лінійний алгоритм поділу навпіл");
print("x^3 - 8*x^2 + 17*x - 10 = 0");
print("x = 1, 2, 5");
print("X0 = [0; 10]");
print(f"{'-'*80}");
x1 = 1;
x2 = 2;
x3 = 5;
e = 10**(-3)

f = lambda x: x**3 - 8*x**2 + 17*x - 10

a = 0;
b = 6;

linear_half(f, a, b, e)
```

```
#@title Алгоритм Мура (для 1 р-ння)
def linear_moore(f, f_prime, start, end, e):
  width = end - start
  if width < e:
    print(f'FOUND X = [{start}; {end}]')
    return

  X0 = interval([start, end])
  mid = (start + end) / 2;
  f_mid = f(mid)

  intervals = np.array([f"[{start}; {end}]"])
  widths = np.array([width])
  middles = np.array([mid])
  f_prime_intervals = np.array([f""])
  u_intervals = np.array([f""])
  intervals_stack = []

  if f_mid == 0.0:
    intervals_stack.append((mid, end))
    intervals_stack.append((start, mid))
  else:
    f_derivative = f_prime(X0)
    U = mid - f_mid/f_derivative
    Xn = U & X0

    intervals = np.append(intervals, f"[{start}; {end}]")
    widths = np.append(widths, width)
    middles = np.append(middles, mid)
    u_intervals = np.append(u_intervals, f"[{U[0].inf}; {U[0].sup}]")
    f_prime_intervals = np.append(f_prime_intervals, f"[{f_derivative[0].inf}; {f_derivative[0].sup}]")

    for x in Xn:
      intervals_stack.append((x.inf, x.sup))
  
  while len(intervals_stack) != 0:
    Xn = interval(intervals_stack.pop())
    start = Xn[0].inf
    end = Xn[0].sup
    width = end - start
    
    if width <= e:
      intervals = np.append(intervals, "Result")
      widths = np.append(widths, "")
      middles = np.append(middles, "")
      u_intervals = np.append(u_intervals, "")
      f_prime_intervals = np.append(f_prime_intervals, f"Found x: [{start}; {end}]")
      continue
    
    mid = (start + end) / 2;
    f_mid = f(mid)
    
    if f_mid == 0.0:
      intervals_stack.append((mid, end))
      intervals_stack.append((start, mid))
      continue
    
    f_derivative = f_prime(Xn)
    U = mid - f_mid/f_derivative
    Xn = U & Xn

    intervals = np.append(intervals, f"[{start}; {end}]")
    widths = np.append(widths, width)
    middles = np.append(middles, mid)
    u_intervals = np.append(u_intervals, f"[{U[0].inf}; {U[0].sup}]")
    f_prime_intervals = np.append(f_prime_intervals, f"[{f_derivative[0].inf}; {f_derivative[0].sup}]")

    for x in Xn:
      intervals_stack.append((x.inf, x.sup))

  pd.set_option('display.max_rows', 999)  
  df = pd.DataFrame({"Xn": intervals, "ω(Xn)": widths, "x̅_n": middles, "U": u_intervals, "F'(Xn)": f_prime_intervals})
  display(df)
  return

#@title Алгоритм Мура (задача для 1 р-ння)
print("Лінійний алгоритм Мура");
print("x^3 - 6*x^2 + 11*x - 6 = 0");
print("1, 2, 3");
print("[0; 1.5]");
print('-'*80);
x1 = 1;
x2 = 2;
x3 = 3;
e = 10**(-6)

f = lambda x: x**3 - 8*x**2 + 17*x - 10
f_prime = lambda x: 3*x**2 - 16*x + 17

a = 0;
b = 1.5;

linear_moore(f, f_prime, a, b, e)
```

```python
from interval import interval
from interval import imath
import pandas as pd
import numpy as np
import math
```

```python
def calculate_width(x_i):
    return sum([x.sup - x.inf for x in x_i])

def calculate_midpoint(x_i):
    return (x_i[0].sup + x_i[0].inf) / 2

def itos(x_i):
    return f"[{x_i[0].inf:9.7f}, {x_i[0].sup:9.7f}]"
```

## Interval division 

```python
def interval_division(f, start, end, tol, tbl):
        X = interval[start, end]
        X_width = calculate_width(X)
        F = f(X)
        mid = calculate_midpoint(X)
        
        if 0 not in F: #Root not found
            tbl.loc[len(tbl)] = [itos(X), itos(F), X_width, mid, '',]
            return
        
        if X_width < e: #Root founded
            tbl.loc[len(tbl)] = [itos(X), itos(F), X_width, mid, 'Root']
            return
        
        tbl.loc[len(tbl)] = [itos(X), itos(F), X_width, mid, '']
        interval_division(f, start, mid, e, tbl)
        interval_division(f, mid, end, e, tbl)
```

```python
func = lambda x: imath.sin(x-1)*(x-2)*(x-3)
e = 1e-2
a = 0.9
b = 2.3
```

```python
result = pd.DataFrame(columns=['Interval', 'Interval extension', 'Width', 'Mid point', 'Root'])
interval_division(func, a, b, e, result)
result
```
